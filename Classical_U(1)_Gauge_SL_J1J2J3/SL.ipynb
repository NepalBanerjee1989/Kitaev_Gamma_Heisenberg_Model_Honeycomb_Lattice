```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# Load your spin data
# ----------------------------
# Replace 'spin_data.txt' with your actual file
# File format: T, i, j, sx, sy, sz
data = np.loadtxt('spin_dynamics')

# Extract lattice indices and spin components
i = data[:, 1].astype(int)
j = data[:, 2].astype(int)
sx = data[:, 3]
sy = data[:, 4]
sz = data[:, 5]
T=data[0,0]
# Determine lattice size
Lx = i.max() + 1
Ly = j.max() + 1

# ----------------------------
# Create 2D arrays for spins
# ----------------------------
Sx = np.zeros((Lx, Ly))
Sy = np.zeros((Lx, Ly))
Sz = np.zeros((Lx, Ly))

for idx in range(len(i)):
    Sx[i[idx], j[idx]] = sx[idx]
    Sy[i[idx], j[idx]] = sy[idx]
    Sz[i[idx], j[idx]] = sz[idx]

# ----------------------------
# Plot in-plane spin components (XY)
# ----------------------------
X, Y = np.meshgrid(np.arange(Lx), np.arange(Ly))

plt.figure(figsize=(8, 6))
plt.quiver(X, Y, Sx.T, Sy.T, Sz.T, cmap='coolwarm', scale=40, pivot='middle')
plt.title('Spin Configuration in Real Space ')
plt.xlabel('$L_x$')
plt.ylabel('$L_y$')
plt.gca().invert_yaxis()  # Optional: match lattice orientation
plt.colorbar(label='Sz')
plt.axis('equal')
plt.title('T='+str(T),fontsize=20) 
plt.savefig('Spin_Config_J1_J2_J3_0.01.png',bbox_inches = 'tight')
plt.show()
```
